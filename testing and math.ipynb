```python
import wave
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

audio_file = 'joe.wav'
def load_waveform():
    # Open the WAV file
    with wave.open(audio_file, 'rb') as wf:
        # Read audio frames
        # n_frames = wf.getnframes()
        sample_rate, audio_data = wavfile.read(audio_file)

        if (len(audio_data.shape) >1):
           audio_data = np.mean(audio_data, axis=1)
        

        # Normalize audio samples to range [-1, 1]
        normalized_samples = audio_data / np.max(np.abs(audio_data))
        
    return sample_rate, normalized_samples

def plot_waveform(waveform, sample_rate):
    # Plot the waveform
    plt.figure(figsize=(12, 6))
    time = np.arange(0, len(waveform)) / sample_rate
    plt.plot(time, waveform)
    plt.title("Waveform")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()


sample_rate, waveform = load_waveform()
plot_waveform(waveform, sample_rate)
```
